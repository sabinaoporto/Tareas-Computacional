## 1. Leemos los archivos

```python
import pandas as pd
```

```python
## con estas lineas de codigo estamos accediendo a los archivos que se encuentran ubicados en la misma carpeta que este 
## jupiter, a fin de leer todas juntas y arrohar el proceso de tener que leer una por una 

import glob as glob

archivos_txt = glob.glob('*.txt')
archivos = {fp.replace('.txt',''):pd.read_csv(fp, sep='\t') for fp in archivos_txt}
```

```python
## visualizamos los datos convertidos en una serie de tablas gracias a la librería usada antes

for rasgo in archivos.keys():
    print(archivos[rasgo])
```

## 2. Creamos diccionarios de rasgos

```python
## en estas lineas estamos creando un diccionario que va a tener 5 entradas principales correspondientes a los 5 rasgos que 
## escogi para trabajar. además, se está creando, dentro de cada uno de esas 5 entradas, multiples entradas más que tienen
## como key los nombres de las lenguas que se registran dentro de cada rasgo y como value la clasificación se acuerdo con este

diccionarios = {}

for rasgo in archivos.keys():
    diccionario_rasgo = archivos[rasgo]
    nombre_valor = dict(zip(diccionario_rasgo['name'],diccionario_rasgo['value']))
    diccionarios[rasgo] = nombre_valor
```

```python
## visualizamos solo uno de los rasgos contenidos dentro del dicc 'diccionario'

diccionarios ['ARGEX10-3 independent reflexive pron']
```

```python
## juntamos todas las lenguas en otro diccionarios con values vacios para poder reorganizar la información del dicc 
## 'diccionarios', para ello usamos un ciclo for y recorremos los 5 rasgos de 'diccionarios' y luego creamos una lista con
## los nombres de las lenguas que usaremos para colocarlas en los keys del nuevo dicc 'rasgos' que estamos creando 

rasgos = {}

for rasgo in diccionarios.keys():
    lenguas = list(diccionarios[rasgo].keys())
    for lengua in lenguas:
        rasgos[lengua] = {}
```

```python
## visualizamos el resultado de lo anterior

rasgos
```

```python
## usando el comando len podemos saber la cantidad de lenguas que estamos abarcando para el análisis de los rasgos

len(rasgos)
```

```python
## con el siguiente código vamos a llenar el dicc 'rasgos', que tiene como keys todas las lenguas, con los rasgos que pueden
## aparecer en cada una de ellas

for rasgo in diccionarios.keys():
    for lengua in rasgos.keys():
        if lengua in diccionarios[rasgo].keys():
            rasgos[lengua][rasgo]=diccionarios[rasgo][lengua]
```

## 3. Distancias

```python
## distancia de Hamming! nos sirve para calcular la diferencia entre una y otras lenguas en cuanto a los rasgos que presentan

def hamming(nombre1,nombre2):
    ## número de rasgos en que las lenguas difieren
    d = 0
    ## número total de rasgos en común
    n = 0
    ## accedemos a los diccionarios de rasgos
    lengua1 = rasgos[nombre1]
    lengua2 = rasgos[nombre2]
    ## contamos con un ciclo for
    for rasgo in lengua1.keys():
        ## nos preguntamos si rasgo pertenece a ambas lenguas
        if rasgo in lengua2.keys():
            ## revisamos si poseen valores distintos
            if lengua1[rasgo] != lengua2[rasgo] and lengua1[rasgo]!='?' and lengua2[rasgo]!='?':
            #if lengua1[rasgo] != lengua2[rasgo]:
                
                ## en este caso sumamos 1 a d
                d += 1.0
            ## número total de rasgos
            n += 1.0
    ## entregamos la división entre el número rasgos en que las lenguas difieren y el número total de rasgos comunes
    return d/n
```

```python
## esta es la distancia entre el matsés y el yaminahua

hamming('Matsés','Yaminahua')
```

```python
## la lengua que se va a comparar es el matses

lengua_comp ='Matsés'
```

```python
distancias = {}
```

```python
for lengua in rasgos.keys():
    distancias[lengua]=hamming(lengua_comp,lengua)
```

```python
distancias_ordenadas = dict(sorted(distancias.items(), key=lambda item: item[1]))
```

Por último, los puntos en colores más transparentes son aquellas lenguas que al ser comparadas con el matsés arrojan valores centrales entre 0 y 1.
