## Preguntas

### 1. ¿Cuántos lemas distintos hay?

```python
## Defina la lista lemas_distintos, que guarda los lemas distintos de la lista lemas

lemas_distintos = []
## use un ciclo for para recorrer la lista lemas. 

for string in lemas:
    ## Usamos un if dentro de un ciclo for para que vaya ingresando a la nueva lista los lemas que no se encuentren ya dentro.
    if string not in lemas_distintos:
        lemas_distintos+=[string]
```

### 3. Defina la lista tokens_codigos, en que cada código está tokenizado por ";"

```python
## use un ciclo for para recorrer la lista codigos. En cada iteración, use split(';') y guarde el resultado en tokens_codigos

tokens_codigos = []

for string in codigos:
    ## cortamos por ;
    tokens_codigos+=[string.split(';')]
```

### 4. ¿Cuál es el número promedio de símbolos usados para describir las inflexiones?

```python
## recorra la lista tokens_codigos con un ciclo for. En cada iteración, guarde el largo de la sublista en inflexiones_totales_usadas.
## Finalmente, divida inflexiones_totales_usadas por el largo de tokens_codigos!

########################## LO QUE HABIA HECHO JAVIER ##############################
## promedio_inflexiones = 0  ########## esta cosa NO ES UN PROMEDIO!!! ES EL TOTAL DE INFLEXIONES que se están usando y que, 
                             ########## luego de muchos pasos :'v, recién nos ayudarán a encontrar el promedio 

######################### COMO LO CAMBIÉ #############################

inflexiones_totales_usadas = 0

## siguiendo los pasos de javier, tenemos que ir midiendo la cantidad de inflexiones (símbolos) usando len y debemos guardar
## esa medida inflexiones_totales_usadas

## entonces.... ## recorra la lista tokens_codigos con un ciclo for:

for item in tokens_codigos:
    
    ## ahora codeamos esto: guarde el largo de la sublista en promedio_inflexiones.
    
    inflexiones_totales_usadas+= len(item)
```

```python
## EL LAST STEP: inalmente, divida promedio_inflexiones por el largo de tokens_codigos!

## primero definimos la lista promedio_inflexiones pa guardar el resultado 

promedio_inflexiones = 0

## y por último, hace la división. OJO: usamos len para la lista tokens codigos a fin de que se pueda operar con números

promedio_inflexiones = inflexiones_totales_usadas/len(tokens_codigos)
```

### 5. Ahora, tokens_codigos es una lista de listas. Defina la lista lista_codigos, que guarda todos los strings usados para describir las inflexiones

```python
lista_codigos = []

## use un ciclo for para recorrer la lista tokens_codigos. En cada iteración, use otro ciclo for, que recorra la sublista, y 
## guarde los strings en lista_codigos

for sublista in tokens_codigos:
    for item in sublista:
        lista_codigos += [item]
```

```python
## Defina la lista simbolos_distintos, que guarda los simbolos distintos de lista_codigos

simbolos_distintos = []
## use un ciclo for para recorrer la lista lista_codigos. 

for simbolo in lista_codigos:
    ## Usamos un if dentro de un ciclo for para que vaya ingresando a la nueva lista los simbolos que no se encuentren dentro
    if simbolo not in simbolos_distintos:
        simbolos_distintos+=[simbolo]
```

### 7. Construya la lista triple_datos en donde cada elemento tiene la forma [lema,inflexión,código]

```python
triple_datos = []

## recorra las listas usando los índices!


######## SO, usamos el código de las evoluciones de los pokemones uwuwu
# for i in indices_1:
#     evolucion = pokemones_1[i]+""+pokemones_2[i]+pokemones_3[i]
#     evolucion_1 += [evolucion]

### la idea de este ciclo for es fusionar el elemento 1 de cada lista (lemas, inflexiones y codigos) y ponerlo en una mini 
### lista dentro de una lista más grande llamada triple_datos

for i in range(len(lemas)):
    fusion = lemas[i] + " " + inflexiones[i] + " " + codigos[i]
    
    ### para que aparezca cada mini lista tenemos que hacer split con la fusión .split
    
    haciendo_el_split = fusion.split()
    
    ### y ahora sí lo metemos todo dentro de una lista grande: triple_datos
    
    triple_datos += [haciendo_el_split]
```
